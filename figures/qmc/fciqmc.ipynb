```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('c2.fcimcstats')
# this is C2, ccpvdz, at 1.2425 Angstrom (equilibrium)
```

```python
Eref = -75.386902377705638
iterations = data[:,0]
shift = data[:,1] + Eref
Eproj = data[:,22]
Etrial = data[:,33]/data[:,34]
population = data[:,4]
```

```python
plt.plot(iterations, shift, label='Shift')
plt.plot(iterations, Eproj, label='Projected Energy')
plt.plot(iterations, Etrial, label='Trial Energy')
plt.xlabel('FCIQMC Iteration')
plt.ylabel('Energy Estimation (Hartree)')
plt.legend()
plt.savefig('c2_example.pdf', bbox_inches='tight')
```

```python
plt.plot(iterations, shift, label='Shift')
plt.plot(iterations, Eproj, label='Projected Energy')
plt.plot(iterations, Etrial, label='Trial Energy')
plt.xlabel('FCIQMC Iteration')
plt.ylabel('Energy Estimation (Hartree)')
plt.legend()
plt.xlim(left=3000)
plt.ylim(bottom=-75.73,top=-75.727)
plt.savefig('c2_example_zoomed.pdf', bbox_inches='tight')
```

```python
plt.plot(iterations, population)
plt.xlabel('FCIQMC Iteration')
plt.ylabel(r'Walker Population, $\sum |c_i|$')
# plt.legend()
plt.savefig('c2_example_plateau.pdf', bbox_inches='tight')
```
