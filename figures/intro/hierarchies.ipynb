```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

x_words = ["STO-3G", "cc-pVDZ", "cc-pVTZ", "..."]
y_words = ["HF", "CISD", "CCSD", "..."]

ax.set_xlim(0, 2)
ax.set_ylim(0, 2)

ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(x_words)
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(y_words)

ax.set_xlabel("Basis Set")
ax.set_ylabel("Method")

ax.annotate("CBS", xy=(1, 0.25), xytext=(2.5, 0.25), verticalalignment="center",
            arrowprops=dict(arrowstyle="<-"))
ax.annotate("FCI", xy=(0.25, 1), xytext=(0.25, 2.5), horizontalalignment="center",
            arrowprops=dict(arrowstyle="<-"))
ax.annotate("Exact", xy=(0.5, 0.5), xytext=(2.5, 2.5), arrowprops=dict(arrowstyle="<-"))

plt.savefig("hierarchies.pdf")
```
