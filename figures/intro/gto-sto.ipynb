```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def slater_function(zeta, R):
    return lambda r: (zeta**3/np.pi)**0.5 * np.exp(-zeta*np.abs(r-R))

def gaussian_function(alpha, R):
    return lambda r: (2*alpha/np.pi)**(3/4) * np.exp(-alpha*np.abs(r-R)**2)
```

```python
gf = gaussian_function
target_sto = slater_function(1.0, 0.0)
sto1g = gf(0.270950, 0.0)
sto3g = lambda r: 0.444635*gf(0.109818, 0.0)(r) + 0.535328*gf(0.405771, 0.0)(r) + 0.154329*gf(2.22766, 0.0)(r)
```

```python
r = np.linspace(0, 4.0, 100)
plt.plot(r, target_sto(r), label="Slater 1s orbital")
plt.plot(r, sto1g(r), '--', label="STO-1G")
plt.plot(r, sto3g(r), '--', label="STO-3G")
plt.legend()
plt.xlabel(r"$r$/Bohr")
plt.ylabel(r"$\phi_\mathrm{1s}$")
plt.xlim(0, 4)
plt.ylim(0, 0.6)
plt.savefig('gto-sto.pdf')
```
