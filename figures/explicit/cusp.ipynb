$|x| = \frac \pi 2 - \frac 4\pi \sum_{n=1}^\infty\frac{\cos[(2n-1)x]}{(2n-1)^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib
```

```python
# x = np.linspace(-np.pi, np.pi, 101)
x = np.linspace(-0.5, 0.5, 101)
y = np.abs(x)
```

```python
def abs_fourier(N, x):
    retval = np.zeros_like(x)
    for i in range(0,N):
        n = i + 1
        retval += np.cos((2*n-1) * x) / (2*n-1)**2
    return np.pi/2 - 4/np.pi * retval
```

```python
# plt.xlim(-np.pi, np.pi)
plt.plot(x, y, 'k-', label=r'$|x|$ (N=$\infty$)')
# plt.plot(x,abs_fourier(0,x))
n_all = [1,2,3,4,5,25]
for n in n_all:
    plt.plot(x,abs_fourier(n,x), label=f'N={n}')
# plt.show()
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$\frac{\pi}{2} - \frac {4}{\pi} \sum_{n=1}^N\frac{\cos[(2n-1)x]}{(2n-1)^2}$')
# plt.ylabel(r'$\frac \pi 2 - \frac 4\pi \sum_{n=1}^\infty\frac{\cos[(2n-1)x]}{(2n-1)^2}$')
plt.savefig('cusp.pdf')
```
