```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5000
# get 1000 random numbers between -1 and 1
x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)
```

```python
# categorise x and y based on if in unit circle
x_in = []
y_in = []
x_out = []
y_out = []
for i in range(len(x)):
    if x[i]**2 + y[i]**2 <= 1:
        x_in.append(x[i])
        y_in.append(y[i])
    else:
        x_out.append(x[i])
        y_out.append(y[i])
```

```python
# Define the unit circle
t = np.linspace(0, 2*np.pi, 100)
xcirc = np.cos(t)
ycirc = np.sin(t)
# make plotting area square
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
# Plot the unit circle
plt.plot(xcirc, ycirc, 'k')
plt.plot(x_in, y_in, '.')
plt.plot(x_out, y_out, '.')
plt.savefig('darts.pdf') # this particular image gives pi ~ 3.1464
```

```python
N_in = len(x_in)
N_out = len(x_out)
N_tot = N_in + N_out
```

```python
# area of circle  = pi * r^2
# area of square = (2r)^2
# area of circle / area of square = pi / 4
approx_pi = 4 * N_in / N_tot
print(approx_pi)
```
